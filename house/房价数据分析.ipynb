```python
def month_mean(month,price):
    sum_=0
    price_mean=[]
    for i in range(len(month)):
        front=sum_
        sum_+=month[i]
        sub_price=price[front:sum_]
        mean=np.mean(sub_price)
        price_mean.append(mean)
    return price_mean
```

```python
max_=feature_42_68.max(axis=0)
min_=feature_42_68.min(axis=0)
print max_
print min_
```

```python
for i in range(m):
    for j in range(n):
        if feature_42_68[i][j]==max_[j] :
            feature_42_68[i][j]=1.0
        else:
            feature_42_68[i][j]=(float(feature_42_68[i][j])-float(min_[j]))/(float(max_[j])-float(min_[j]))
```
